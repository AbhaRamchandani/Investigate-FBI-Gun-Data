### Research Question 1: Investigate Gun Purchase trends.

> These statistics are based on the number of firearm background checks initiated through the NICS. They do not represent the number of firearms sold. 

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
```

```python
## Find total number of firearm background checks per year
df_bkgchk_yr = df_gun[['month','totals']].groupby(df_gun['month']).sum()
df_bkgchk_yr.head(5)
```

```python
## Plot background checks vs time (year)
plt.figure(figsize=(16, 8))   

ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.grid(which='major', axis='both', linestyle='--', alpha=0.5, color="#3F5F7F")

plt.yticks(fontsize=14)    
plt.xticks(fontsize=14)   

plt.ylim(0, 3500000)    
plt.xlim(datetime.date(1998, 11, 1), datetime.date(2018, 1, 1))

plt.plot(df_bkgchk_yr, color="#3F5F7F")

plt.title("Firearm background checks USA (1998-2017)",
          fontsize=18,
          ha="center")

plt.xlabel("Year",
           fontsize=14,
           labelpad=20)

plt.ylabel("Number of Background Checks", 
           fontsize=14,
           labelpad=20)

plt.show()
```

> Clearly, there has been an overall increase in the number of firearm background checks over the years. 

> Let's check if there is a seasonal trend!

```python
## Find total number of firearm background checks per year
df_bkgchk_m = df_bkgchk_yr.reset_index()
df_bkgchk_m = df_bkgchk_m.groupby(df_bkgchk_m['month'].dt.strftime('%B'))['totals'].sum().reset_index()

## Sorting (by month) https://stackoverflow.com/a/48043319
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

df_bkgchk_m['month'] = pd.Categorical(df_bkgchk_m['month'], categories=months, ordered=True)
df_bkgchk_m.sort_values(by='month',ascending=True,inplace=True)
df_bkgchk_m.reset_index(drop=True, inplace=True)
df_bkgchk_m
```

```python
## Plot background checks by month
plt.figure(figsize=(14,8)) 

ax = plt.subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)    
ax.grid(which='major', axis='y', linestyle='--', alpha=0.5, color="#3F5F7F")

plt.xticks(fontsize=14, rotation=75)
plt.yticks(fontsize=14) 

plt.tick_params(axis="both", which="both", bottom=False, top=False, labelbottom=True, left=False, right=False, labelleft=True) 

plt.ylim(0, 35000000)    

bar_list = plt.bar(df_bkgchk_m['month'], df_bkgchk_m['totals'], color="#3F5F7F", edgecolor="k")

for a,b in zip(df_bkgchk_m['month'], df_bkgchk_m['totals']):
    plt.text(a, b + 250000, str(b), ha='center', va='bottom', fontsize=11)

plt.title("Firearm background checks by month, USA (1998-2017)",
          fontsize=18,
          ha="center")

plt.xlabel("Month",
           fontsize=14,
           labelpad=20)

plt.ylabel("Count", 
           fontsize=14,
           labelpad=20)
        
plt.show()
```

> Fire arm sales are highest in the month of March. This cannot be attributed to any specific event at the moment, based on the data we have.

### Research Question 2: Which states has had highest growth in gun registrations from since 2000?


```python
## Udacity's step description!

# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
```

> We will look at years from 2000 to 2016 since we do not have complete data for 2017.

```python
## Total number of firearm background checks per year by state 
df_bkgchk_state = df_gun[['month','totals','state']]
df_bkgchk_state = df_bkgchk_state.groupby([df_bkgchk_state['state'], df_bkgchk_state['month'].dt.strftime('%Y')])['totals'].sum().reset_index()
df_bkgchk_state.columns = ['state', 'year', 'totals']

## We will drop any data from 1998, 1999, 2017
df_bkgchk_state = df_bkgchk_state[(df_bkgchk_state['year'] != '1998') & (df_bkgchk_state['year'] != '1999') & (df_bkgchk_state['year'] != '2017')]
df_bkgchk_state.head()
```

```python
## Calculate total number of firearm background checks for each state from 2000 to 2017.
## Results will be stored in a new dataframe.

growth = []

for state in df_bkgchk_state['state'].unique():
    first_year = df_bkgchk_state[(df_bkgchk_state['state'] == state) & (df_bkgchk_state['year'] == '2000')]['totals'].iloc[0]
    last_year = df_bkgchk_state[(df_bkgchk_state['state'] == state) & (df_bkgchk_state['year'] == '2016')]['totals'].iloc[0]
    nbr_change = last_year - first_year
    pct_change = int(round((last_year/first_year) * 100))
    
    result = pd.DataFrame({
        "state": state,
        "first_year": first_year, 
        "last_year": last_year,
        "nbr_change": nbr_change, 
        "pct_change": pct_change
    }, index=[0])

    growth.append(result)
            
    print('{}: {}% '.format(state, pct_change))
```

```python
## New dataframe for results
df_growth = pd.concat(growth, axis=0, ignore_index=True)
df_growth.columns = ['state', 'totals_2000', 'totals_2016', 'nbr_change', 'pct_change']
df_growth.set_index('state', inplace=True)
df_growth.head()
```

```python
## Plot percent growth by state
plt.figure(figsize=(20,10)) 

ax = plt.subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)    
ax.grid(which='major', axis='y', linestyle='--', alpha=0.5, color="#3F5F7F")

plt.xticks(fontsize=14, rotation=90)
plt.yticks(range(0, 1601, 200), [str(x) + "%" for x in range(0, 1601, 200)], fontsize=14) 

plt.tick_params(axis="both", which="both", bottom=False, top=False,
                labelbottom=True, left=False, right=False, labelleft=True) 

plt.ylim(0, 1650)    
plt.xlim(-1, 50)  

bar_list = plt.bar(df_growth.index, df_growth['pct_change'], color="#3F5F7F", edgecolor="k")

for a,b in zip(df_growth.index, df_growth['pct_change']):
    plt.text(a, b, str(round(b,2)), ha='center', va='bottom', fontsize=12)

plt.title("Percentage increase in the number of firearm background checks by State, USA, (2000 to 2016)",
          fontsize=18,
          ha="center")

plt.xlabel("State",
           fontsize=14,
           labelpad=20)

plt.ylabel("Increase %", 
           fontsize=14,
           labelpad=20)
        
plt.show()
```
